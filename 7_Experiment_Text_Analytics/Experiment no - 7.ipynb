```python
def tokeniztion(text):
    tokens = word_tokenize(text)
    return tokens
```

```python
df['content_tokens'] = df['content'].apply(tokeniztion)
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stemmer = PorterStemmer()
```

```python
nltk.download('punkt')
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [re.sub(r'[^\w\s]' , '' , word).lower() for word in tokens]
    tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    return tokens
```

```python
df['tokens_content'] = df['content'].apply(preprocess_text)
```

```python
from textblob import TextBlob
```

```python
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiments = blob.sentiment.polarity
    return sentiments
```

```python
df['sentiments'] = df['content'].apply(analyze_sentiment)
```
