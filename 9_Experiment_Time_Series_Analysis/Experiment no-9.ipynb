```python
df = pd.read_csv('/content/drive/MyDrive/Data Analytics/DA LAB POE/9_Experiment_Time_Series_Analysis/time_series.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['number_sold'] = pd.to_numeric(df['number_sold'])
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['number_sold'], label='')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```
