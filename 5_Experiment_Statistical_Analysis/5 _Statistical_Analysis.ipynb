```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Analytics/DA LAB POE/Experiment_5/synthetic_data.csv')
df.head()
```

```python
from scipy.stats import ttest_ind

# Filter the data for the two education levels
bachelor_income = df[df['education_level'] == 'Bachelor']['income']
master_income = df[df['education_level'] == 'Master']['income']

# Perform the t-test
t_stat, p_value = ttest_ind(bachelor_income, master_income)

t_stat, p_value
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['education_level'], df['satisfaction_score'])

# Perform the chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

chi2_stat, p_value
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Encode categorical variable
df['education_level_encoded'] = df['education_level'].astype('category').cat.codes

# Define the model
model = ols('income ~ age + education_level_encoded', data=df).fit()

# Summary of the model
model_summary = model.summary()
model_summary
```
